```python
cmt.isna().sum()
```

```python
#.isna().sum() used to know the missing value
```

```python
cmt.dropna(inplace=True)
```

```python
#.dropna() used for dropping the missing value 
#inplace=True used for permanent change
```

```python
polarity=[]

for i in cmt['comment_text']:
    polarity.append(TextBlob(i).sentiment.polarity)
```

```python
#here the for loop used for iteration in column 'comment_text'
#through which polarity of each comment can be known.
```

```python
cmt['polarity']=polarity
```

```python
#by this syntax new column 'polarity' can be added.
```

```python
from wordcloud import WordCloud,STOPWORDS
```

```python
#here is wordcloud library is being imported
#used for visual representation of text data
```

```python
stpwrds=set(STOPWORDS)
```

```python
#stopwords are those words that do not have any useful information 
#to decide in which category a text should be classified.
```

```python
cmt_pstv['comment_text']
```

```python
#here by above syntax ..all the comments get printed
```

```python
''.join(cmt_pstv['comment_text'])
```

```python
#by using space and inverted commas
#join
#to join all the comments which are present in dataset
```

```python
ttl_cmmt=''.join(cmt_pstv['comment_text'])
```

```python
#here we have given name as ttl_cmmt to jointed comments
```

```python
wrdcld=WordCloud(width=1200,height=600,stopwords=stpwrds).generate(ttl_cmmt)
```

```python
#here we specify the size of word
#using stopwords ,we can ommit the useless word in comment
```

```python
plt.figure(figsize=(20,8))
plt.imshow(wrdcld)
plt.axis('off')
```

```python
#here we used figsize for size of whole poster above.
```

```python
#here the conclusion is that by looking the size of word 
#in poster we can predict that the words are written most of the time in comment section.
```

```python
#Now the same concept will be applied on the negative comments
```

```python
cmt_ngtv=cmt[cmt['polarity']==-1]
```

```python
stpwrds=set(STOPWORDS)
```

```python
cmt_ngtv['comment_text']
```

```python
ttll_cmmtt=''.join(cmt_ngtv['comment_text'])
```

```python
wrdclld=WordCloud(width=1200,height=600,stopwords=stpwrds).generate(ttll_cmmtt)
```

```python
plt.figure(figsize=(20,8))
plt.imshow(wrdclld)
plt.axis('off')
```

```python
#here is the conclusion that word 'boring'
#most bigger among all negative words
#means it is written most of the time in comment section of you tube.
```
